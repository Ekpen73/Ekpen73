```python
# data normalization with sklearn
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing
from sklearn.preprocessing import scale
# transform testing dataabs
X=df1.iloc[:,:-1]# Create a vector of independent vairable
y=df1.iloc[:,-1]
norm = MinMaxScaler().fit_transform(X)
norm=pd.DataFrame(norm,columns=['fixed acidity','volatile acidity','citric acid','residual_sugar','free sulfur dioxide','Cloride','total sulfur dioxide','density','pH','sulphates','alcohol'])#'quality'])             
norm
```

### Using Scree Plot to Vizualize the data set FOR DIMENSIONALITY REDUCTION

```python
from factor_analyzer import FactorAnalyzer
from sklearn.decomposition import FactorAnalysis
# Create factor analysis object and perform factor analysis
fa = FactorAnalyzer(n_factors=3, rotation="varimax")
fa.fit(norm)
ev,_ = fa.get_eigenvalues()
plt.scatter(range(1,norm.shape[1]+1),ev)
plt.plot(range(1,norm.shape[1]+1),ev)
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigen Value')
plt.grid()
```

The plot shows the factors and eigenvalues.  The eigenvalue are the amount of variance the factor explains and this is usually values greater than 1. Using 3, the eigen value will be 1.5, and there are still values greater than 1. At factor 4, the eigen value is at optimal and it drops below 1 at 5.  The projection of 4 dimension will capture the important details of the original dataset.

### Principal Component Analysis

```python
from sklearn.decomposition import PCA
from sklearn import decomposition #PCA
#Build the model
pca=decomposition.PCA(n_components=6) 
#Reduced the data
features=pca.fit_transform(norm)
#pd.DataFrame(features,columns=['PCA1','PCA2','PCA3']) # choose 3 dimension reduction
#pd.DataFrame(features,columns=['P1','P2','P3','P4']) # choose 4 dimension reductio
#pd.DataFrame(features,columns=['P1','P2','P3','P4','P5'])#,'P9','P10','P11','P12','P13','P14','P15'))
pd.DataFrame(features,columns=['P1','P2','P3','P4','P5','P6'])
```

```python
pca.explained_variance_ratio_
```

In the above code inputing different component gives the following:  6 dimensions accounts for 87%,  5 dimensions accounts for 82%,  4 dimensions accounts for 77% and 3 dimensions accounts for 70%.  From the cummulative graph below, at 70% the variance started increasing at a decreasing rate and after 77%, it increases at a more decreased rate.  We can take around 4 as the optimal, so projecting at this dimension will capture almost all details.

```python
cum_explained_var = []
for i in range(0, len(pca.explained_variance_ratio_)):
    if i == 0:
        cum_explained_var.append(pca.explained_variance_ratio_[i])
    else:
        cum_explained_var.append(pca.explained_variance_ratio_[i] + 
                                 cum_explained_var[i-1])

print(cum_explained_var)
fig, ax = plt.subplots()
ax.plot(cum_explained_var)
```
