```python
# import required libraries
import numpy as np 
import matplotlib.pyplot as plt  
from sklearn.cluster import KMeans
%matplotlib inline
import seaborn as sns; sns.set()
```

#### Create the Model to process the data

```python
from sklearn.cluster import KMeans
Kmean = KMeans(n_clusters=3)
km_labels = Kmean.fit(df[['X','Y']])
km_labels
```

### Find the Centroid of the data

```python
#assign a cluster to each datapoint

km_labels = Kmean.fit_predict(df[['X','Y']])

#retrieve unique cluster centers
centroids = Kmean.cluster_centers_

# Create scatter plot for dataset
plt.scatter(df['X'],df['Y'], c=km_labels,cmap='jet', marker='+')

# This code identifies the centroids
plt.scatter(centroids[:, 0], centroids[:, 1],c='red', s=100)

# Assigns labelling to the graph
plt.title ('Scater Plot with Three Cluster', size ='20')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

## Applying the Gaussian Mixture Model

```python
# import the GMM from python library
import pandas as pd
from sklearn.mixture import GaussianMixture

df = pd.read_csv('cluster1(1).csv', names = ['X','Y'])

# training gaussian mixture model 
gmm = GaussianMixture(n_components=3)
gmm.fit(df)

#predictions from GMM
pred = gmm.predict(df)
frame = pd.DataFrame(df)
frame['cluster'] = pred

color=['blue','green','red']
for k in range(0,3):
    df = frame[frame["cluster"]==k]
    plt.scatter(df["X"],df["Y"],c=color[k], marker='+')
    
plt.title ('Graph of GMM Clusters', size ='20')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Justification of choice of Implementation

The K-means graph above; in terms of clustering of data, is used for convergence points where we obtained a general sectioning.  While the Gaussian Mixture Model appears to represent the data in a normalized form similar to general pattern recognition to which the co-relation belongs.

## Method 1: Using Elbow method 


The elbow uses the sum of Squared error and assume that as the number of clusters increases, each single data point will have a cluster center that is closer to it, resulting in a decrease in the sum of squared error.  And by increasing the number of clusters, we get a less decrease in the sum of squared error.

```python
# Elbow method to minimize WSS(within-cluster sum of square)
Sum_of_squared_distances = []
# Using a loop to run K-Means for a range of clusters and collecting the square of distance
K = range (1,12)
for k in K:
    km = KMeans(n_clusters = k)
    km = km.fit(df)
    Sum_of_squared_distances.append(km.inertia_) 
```

```python
#Scatter plot to visualize the elbow
fig = plt.figure(figsize =(8,8))
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')

# The optimal K is determined at 4. This is shown at the deep end of the elbow after which the slope becomes more flattening.
number_of_clusters = 4
plt.axvline(x = number_of_clusters, linewidth=2, color ='blue', linestyle = '--')
plt.show()
```
